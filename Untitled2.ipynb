```python
import os
import time
import random
```

```python
dataset = {}
path = "/content/dataset/"
files= os.listdir(path)
for file in files:
    if not os.path.isdir(file) and not file[0] == '.': # 跳过隐藏文件和文件夹
        f = open(path+"/"+file, 'r',  encoding='UTF-8'); # 打开文件
        for line in f.readlines():
            dataset[line] = file[:-4]
```

```python
name = {'LX':0,'MY':1,'QZS':2,'WXB':3,'ZAL':4}
for (k,v) in list(dataset.items()):
  dataset[k] = name[v]
```

```python
import random


# 将字典转换为列表
items = list(dataset.items())

# 打乱列表
random.shuffle(items)

# 将列表转换回字典
shuffled_dict = dict(items)

for (k,v) in  list(shuffled_dict.items())[5000:5010]:
    print(k,'---',v)
```

```python
num = len(shuffled_dict)
i= 0
train_data = {}
eval_data = {}
for k ,v in shuffled_dict.items():
  if i <= int(num*0.8):
    train_data[k] = v
  else:
    eval_data[k] = v
  i += 1

```

```python
for k ,v in list(eval_data.items())[:10]:
  print(k,v)
```

```python
name_zh = {'LX': '鲁迅', 'MY':'莫言' , 'QZS':'钱钟书' ,'WXB':'王小波' ,'ZAL':'张爱玲'}
for (k,v) in  list(dataset.items())[5000:5010]:
    print(k,'---',v)
```

```python
train_dicts = {'text' : [k for k,_ in train_data.items()], 'labels': [v for _,v in train_data.items()]}
eval_dicts = {'text' : [k for k,_ in eval_data.items()], 'labels': [v for _,v in eval_data.items()]}
```

```python
from datasets import Dataset
transformers_dataset1 = Dataset.from_dict(train_dicts)
transformers_dataset2 = Dataset.from_dict(eval_dicts)
```

```python
from transformers import AutoConfig  # 导入包
checkpoint = "hfl/chinese-bert-wwm-ext"  # 选择模型
config = AutoConfig.from_pretrained(checkpoint)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
train_dataset = transformers_dataset1.map(lambda example: tokenizer(example['text']), batched=True)
eval_dataset = transformers_dataset2.map(lambda example: tokenizer(example['text']), batched=True)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
import evaluate

accuracy = evaluate.load("accuracy")
```

```python
id2label = {0: "LX", 1: "MY",2:'QZS',3:'WXB',4:'ZAL'}
label2id = {"LX": 0, "MY": 1,'QZS':2,'WXB':3,'ZAL':4}
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

model = AutoModelForSequenceClassification.from_pretrained(
    checkpoint, num_labels=5, id2label=id2label, label2id=label2id
)
```

```python
import numpy as np


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
training_args = TrainingArguments(
    output_dir="/content/my_awesome_model",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=2,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
```
